```python
import os 
import nltk
```

```python
file_name = []
dir_path = r'/content/gdrive/MyDrive/workk/dataset/'
count = 0
for path in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, path)): 
        file_name.append(path)
        count += 1
print('File count:', count)
print(file_name)
```

```python
container = ""
for i in range(count):
  try :
    with open('/content/gdrive/MyDrive/workk/dataset/{0}'.format(file_name[i])) as f:
      contents = f.read()
      
      container += str(contents)
      container = container.replace("\n",'').replace(":",'').replace("*",'')
  except Exception as e:
    
    pass
```

```python
print(len(container))
```

```python
con = container.split()
for i in range(500):
  print(con[i], sep = ' ', end=' ')
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(container)
tokens
```

```python
len(tokens)
```

```python
from nltk.probability import FreqDist
freqdist= FreqDist()
```

```python
for j in tokens:
  freqdist[j.lower()] += 1
freqdist
```

```python
head = freqdist.most_common(10)
head
```

```python
#blank tokenizer paragraph
from nltk.tokenize import blankline_tokenize
blank_token = blankline_tokenize(container)
len(blank_token)
```

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python

tokens_bigram = list(nltk.bigrams(tokens))
tokens_bigram
```

```python
tokens_trigrams = list(nltk.trigrams(tokens))
tokens_trigrams
```

```python
#let n =10
tokens_ngrams = list(nltk.ngrams(tokens, 10))
tokens_ngrams
```

```python
#stemming 
from nltk.stem import PorterStemmer, LancasterStemmer

wordstem = ["give", "giving", "given", "gave"]

pst = PorterStemmer()
lst = LancasterStemmer()

for k in wordstem:
  print(k + ":" +pst.stem(k))
  print(k + ":" +lst.stem(k)) 
```

```python
#lemmatization
from nltk.stem import wordnet, WordNetLemmatizer
word_len = WordNetLemmatizer()
for k in wordstem:
  word_len.lemmatize(k)
```
